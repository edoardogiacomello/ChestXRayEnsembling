```python
import pandas as pd
import os
import glob
```

```python
fold = sorted(glob.glob('*/'))
files = tuple(sorted(glob.glob(f + '*.csv')) for f in fold)
```

```python
aggregation = pd.DataFrame()
for condcsv, basecsv in zip(files[0], files[1]):
    if 'single_model' in condcsv:
        continue
    pdcond = pd.read_csv(condcsv, index_col = 0)
    pdcond['method'] = os.path.basename(condcsv).replace('.csv', '')
    pdcond = pdcond.pivot(columns='Label', index='method')
    pdcond['tta'] = 'tta' in condcsv
    pdcond['folder'] = condcsv.split('/')[0]
    
    pdbase = pd.read_csv(basecsv, index_col = 0)
    pdbase['method'] = os.path.basename(basecsv).replace('.csv', '')
    pdbase = pdbase.pivot(columns='Label', index='method')
    pdbase['tta'] = 'tta' in basecsv
    pdbase['folder'] = basecsv.split('/')[0]
    
    aggregation = aggregation.append(pdcond)
    aggregation = aggregation.append(pdbase)
```
